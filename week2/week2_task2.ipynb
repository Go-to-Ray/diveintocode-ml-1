# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N1 = 500
ave1 = [-3,0]
cov1 = [[1.0,0.8],[0.8,1.0]]

#乱数のシード値を固定
np.random.seed(0)

x1_y1 = np.random.multivariate_normal(ave1,cov1,N1)
#転置行列を用いてx1,y1に代入(スライスでも可)
x1,y1 = np.random.multivariate_normal(ave1,cov1,N1).T
#x1 = x1_y1[:,0]
#y1 = x1_y1[:,1]

print("生成された乱数 :\n{}".format(x1_y1))
#print("生成された乱数 :{}\n".format(np.random.multivariate_normal(ave1,cov1,N1)))
```

# 【問題2】散布図による可視化

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
plt.scatter(x1,y1,color = "g",marker = "x",label  ="N1")

plt.xlabel("x",size = 20)
plt.ylabel("y",size = 20)

plt.show()
```

# 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

N2 = 500
ave2 = [0,-3]
cov2 = [[1.0,0.8],[0.8,1.0]]

np.random.seed(0)
x2_y2 = np.random.multivariate_normal(ave2,cov2,N2)
x2,y2 = x2_y2.T
#print("生成された乱数 x2:\n{}\n\n生成された乱数 y2:\n{}".format(x2,y2))

plt.scatter(x1,y1,color = "r",marker = "o",label = "$0$")
plt.scatter(x2,y2,color = "b",marker = "o",label = "$1$")
plt.title("scatter",size = 15)
plt.xlabel("x",size = 15)
plt.ylabel("y",size = 15)
plt.legend()
plt.show()
```

データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
a = np.array([[1.,2,3],[4,5,6]])
b = a*10
np.concatenate([a,b])
```

```python
join_list = np.concatenate((x1_y1,x2_y2))
print("結合後のndarrayのshapeは{}".format(join_list.shape))
```

# 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

```python
N1 = len(x1_y1)
N2 = len(x2_y2)

#要素が0と1の、ラベル付け用のndarrayを生成
zero_one_list = np.concatenate((np.zeros(N1),np.ones(N2))).reshape(1000,1)
#結合
labeled_list = np.concatenate((join_list,zero_one_list),1)
print(labeled_list)
```
